# Knapsack-Problem

In this problem we think us a bagpack and there are given items, which have a weight and a use-value.<br>
We try now to find the best choice of items, so that the weight is as small as possible and the usage is as high as possible.

The items are: (2, 10), (3, 15), (5, 25), (7, 30), (1, 5)<br>
On the first number is the weight and the right number represents the use-value of the item.

The capacity has a given weight limit.

```python
# VARIABLES -> change me
amount_of_items = 30
knapsack_capacity = 20
```

```python
import random

items = []
for i in range(amount_of_items):
    items += [(random.randint(0, knapsack_capacity), random.randint(0, 50))]
items
```

I choose as representation a dictionary, where every item has one index as key and the value indicates if the item is used.

-> The genetic class need a dictionary solution representation

```python
parameters = dict()
for i in range(len(items)):
    parameters[i] = 0
parameters
```

```python
from genetic_algorithm import GA, get_random

class Knapsack_GA(GA):

    def calculate_fitness(self, kwargs, params):
        items = kwargs["items"]
        weight = 0
        usage = 0
        for item_idx, used in params.items():
            if used == 1:
                weight += items[item_idx][0]
                usage += items[item_idx][1]

        res = 0

        if weight > kwargs["knapsack_capacity"]:
            #res += -len(items)*1000000000
            return float('-inf')
        
        # weight is better if smaller + usage is better if bigger
        res += -weight + usage
        
        return res


    def get_random_param_value(self, param_key):
        return get_random(0, 1)
        # items = []
        # total_weight = 0
        # while True:
        #     while rand_idx in items:
        #         rand_idx = get_random(0, len(items))
        #     rand_item = items[rand_idx]
        #     if total_weight+rand_item[0] > knapsack_capacity:
        #         break
        #     else:
        #         items += [rand_idx]
        # return items
```

```python
optimizer = Knapsack_GA(generations=200, population_size=500, mutation_rate=0.5, list_of_params=parameters)
```

add predefined solutions

```python
for limit in [0, 1, 2, 3, 4]:
    solution = dict()
    counter = 0
    for key, value in parameters.items():
        if counter <= limit and get_random() > 0.8:
            solution[key] = 1
            counter += 1
        else:
            solution[key] = value
    optimizer.add_initial_solution(solution)
```

```python
best_params, best_fitness, log = optimizer.optimize(items=items, knapsack_capacity=knapsack_capacity)
print(best_params, best_fitness);
```
